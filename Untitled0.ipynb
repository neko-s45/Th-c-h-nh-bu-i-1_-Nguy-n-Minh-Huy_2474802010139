# BÀI THỰC HÀNH 1: XỬ LÝ DỮ LIỆU

```python
import pandas as pd
import numpy as np

# 3. Tải dữ liệu và in 10 dòng đầu/cuối
df = pd.read_csv('dulieuxettuyendaihoc.csv')
display(df.head(10))
display(df.tail(10))

# 4. Xử lý dữ liệu thiếu cho cột 'DT' (Dân tộc)
df['DT'] = df['DT'].fillna(0)

# 5. Xử lý dữ liệu thiếu cho 'T1' bằng phương pháp Mean
df['T1'] = df['T1'].fillna(df['T1'].mean())

# 6. Xử lý dữ liệu thiếu cho các biến điểm số còn lại
score_cols = ['L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1', 'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2', 'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6', 'DH1', 'DH2', 'DH3']
for col in score_cols:
    df[col] = df[col].fillna(df[col].mean())# 7. Tạo các biến TBM1, TBM2, TBM3
# Công thức: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
for i in [1, 2, 6]:
    suffix = str(i) if i != 6 else '3'
    year = str(i)
    df[f'TBM{suffix}'] = (df[f'T{year}']*2 + df[f'L{year}'] + df[f'H{year}'] + df[f'S{year}'] + df[f'V{year}']*2 + df[f'X{year}'] + df[f'D{year}'] + df[f'N{year}']) / 10

# 8. Tạo biến xếp loại XL1, XL2, XL3
def classify(tbm):
    if tbm < 5.0: return 'Y'
    elif tbm < 6.5: return 'TB'
    elif tbm < 8.0: return 'K'
    elif tbm < 9.0: return 'G'
    else: return 'XS'

for i in range(1, 4):
    df[f'XL{i}'] = df[f'TBM{i}'].apply(classify)

# 9. Chuyển điểm sang thang điểm 4 (Min-Max Normalization)
for i in range(1, 4):
    df[f'US_TBM{i}'] = (df[f'TBM{i}'] / 10) * 4
# 10. Tạo biến kết quả xét tuyển (KQXT)
def check_admission(row):
    kt = row['KT']
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']
    if kt in ['A', 'A1']:
        score = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B':
        score = (dh1 + dh2*2 + dh3) / 4
    else:
        score = (dh1 + dh2 + dh3) / 3
    return 1 if score >= 5.0 else 0

df['KQXT'] = df.apply(check_admission, axis=1)

# 11. Lưu file
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
print('Đã lưu file processed_dulieuxettuyendaihoc.csv')
display(df.head())
```
