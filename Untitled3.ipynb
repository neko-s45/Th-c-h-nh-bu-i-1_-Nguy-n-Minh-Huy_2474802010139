```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Giả sử dữ liệu đã được load vào biến df
# Nếu chưa có, bạn cần load dữ liệu: df = pd.read_csv('path_to_your_file.csv')

# --- PHẦN 1: THỐNG KÊ DỮ LIỆU ---

# 1. Sắp xếp DH1 tăng dần
df_sorted_dh1 = df.sort_values(by='DH1')

# 2. Sắp xếp DH2 tăng dần theo nhóm giới tính
df_sorted_dh2_gt = df.sort_values(by=['GT', 'DH2'])

# Định nghĩa các hàm phân vị cho pivot table
def q1(x): return x.quantile(0.25)
def q2(x): return x.quantile(0.50)
def q3(x): return x.quantile(0.75)

stats_list = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q2, q3]

# 3. Pivot-table DH1 theo KT
pivot_kt = df.groupby('KT')['DH1'].agg(stats_list)

# 4. Pivot-table DH1 theo KT và KV
pivot_kt_kv = df.groupby(['KT', 'KV'])['DH1'].agg(stats_list)

# 5. Pivot-table DH1 theo KT, KV và DT
pivot_kt_kv_dt = df.groupby(['KT', 'KV', 'DT'])['DH1'].agg(stats_list)

# --- PHẦN 2: TRÌNH BÀY DỮ LIỆU ---

# 1. Biểu đồ tần số và tần suất cho GT
gt_counts = df['GT'].value_counts()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
gt_counts.plot(kind='bar', ax=ax1, title='Tần số Giới tính')
gt_counts.plot(kind='pie', ax=ax2, autopct='%1.1f%%', title='Tần suất Giới tính')
plt.show()

# 4. Trình bày dữ liệu KV với Nam (M), Dân tộc Kinh (giả sử mã là 1), thỏa điểm sàn
cond_p2_4 = (df['GT'] == 'M') & (df['DT'] == 1) & (df['DH1'] >= 5) & (df['DH2'] >= 4) & (df['DH3'] >= 4)
print("Thống kê KV cho Nam, dân tộc Kinh, đạt điểm sàn:")
display(df[cond_p2_4]['KV'].value_counts())

# --- PHẦN 4: TRỰC QUAN HÓA DỮ LIỆU NÂNG CAO ---

# 2. Tạo biến phanlopt1
def phan_lop_t1(t):
    if t < 5: return 'k'
    elif t < 7: return 'tb'
    elif t < 8: return 'kh'
    else: return 'g'

df['phanlopt1'] = df['T1'].apply(phan_lop_t1)

# Biểu đồ đường Simple cho T1 (ví dụ 10 dòng đầu)
plt.figure(figsize=(10,4))
plt.plot(df['T1'].head(20), marker='o')
plt.title('Biểu đồ đường Simple cho T1 (20 mẫu đầu)')
plt.show()

# --- PHẦN 5: MÔ TẢ DỮ LIỆU VÀ KHẢO SÁT PHÂN PHỐI ---

# 1. Khảo sát T1
print("--- Thống kê mô tả chi tiết cho T1 ---")
print(df['T1'].describe())
print(f"Skewness: {df['T1'].skew():.4f}")
print(f"Kurtosis: {df['T1'].kurtosis():.4f}")

# Box-Plot và Histogram/QQ-Plot cho T1
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['T1'], color='lightgreen')
plt.title('Box-Plot của điểm T1')
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['T1'], kde=True, ax=ax1, color='skyblue')
stats.probplot(df['T1'], dist='norm', plot=ax2)
plt.show()

# 2. Khảo sát T1 theo nhóm phanlopt1
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='phanlopt1', y='T1', order=['k', 'tb', 'kh', 'g'])
plt.title('So sánh phân phối T1 theo từng nhóm phân lớp')
plt.show()

# 3 & 4. Tương quan DH1 và T1
corr_dh1_t1 = df[['DH1', 'T1']].corr().iloc[0, 1]
print(f"Hệ số tương quan giữa DH1 và T1: {corr_dh1_t1:.4f}")

sns.lmplot(data=df, x='T1', y='DH1', hue='KV', height=6)
plt.title('Tương quan DH1 và T1 phân theo Khu Vực')
plt.show()

# 5. Ma trận tương quan giữa DH1, DH2, DH3
corr_matrix = df[['DH1', 'DH2', 'DH3']].corr()
display(corr_matrix)
sns.pairplot(df[['DH1', 'DH2', 'DH3']])
plt.show()
```
