```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Load dữ liệu
df = pd.read_csv('/content/processed_dulieuxettuyendaihoc.csv')

# Định nghĩa hàm thống kê
def q1(x): return x.quantile(0.25)
def q2(x): return x.quantile(0.50)
def q3(x): return x.quantile(0.75)
stats_list = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q2, q3]

print("--- PHẦN 1: THỐNG KÊ ---")
display(df.groupby('KT')['DH1'].agg(stats_list))
```

```python
print("--- PHẦN 2 & 3: TRÌNH BÀY DỮ LIỆU ---")
# 1. Biểu đồ GT
gt_counts = df['GT'].value_counts()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
gt_counts.plot(kind='bar', ax=ax1, color=['skyblue','pink'], title='Tần số GT')
gt_counts.plot(kind='pie', ax=ax2, autopct='%1.1f%%', title='Tần suất GT')
plt.show()

# 4. KV cho Nam, Kinh, đạt điểm sàn (Giả sử GT='M', DT=1)
cond = (df['GT'] == 'M') & (df['DT'] == 1.0) & (df['DH1'] >= 5) & (df['DH2'] >= 4) & (df['DH3'] >= 4)
print("Thống kê KV cho nhóm Nam, Kinh, đạt điểm sàn:")
display(df[cond]['KV'].value_counts())
```

```python
print("--- PHẦN 4 & 5: NÂNG CAO & PHÂN PHỐI ---")
# Phân lớp T1
df['phanlopt1'] = pd.cut(df['T1'], bins=[0, 5, 7, 8, 10], labels=['k', 'tb', 'kh', 'g'], right=False)

# Khảo sát T1
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['T1'], kde=True, ax=ax1)
stats.probplot(df['T1'], dist='norm', plot=ax2)
plt.show()

# Tương quan DH1, DH2, DH3
sns.heatmap(df[['DH1', 'DH2', 'DH3']].corr(), annot=True, cmap='YlGnBu')
plt.title('Ma trận tương quan điểm thi DH')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# 1. Tải dữ liệu
df = pd.read_csv('/content/processed_dulieuxettuyendaihoc.csv')

# Định nghĩa hàm phân vị
def q1(x): return x.quantile(0.25)
def q2(x): return x.quantile(0.50)
def q3(x): return x.quantile(0.75)
stats_list = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q2, q3]

print("Dữ liệu đã được tải và chuẩn bị sẵn sàng.")
```

### Phần 1: Thống kê dữ liệu

```python
# --- PHẦN 1: THỐNG KÊ DỮ LIỆU ---

# 1. Sắp xếp DH1 tăng dần
df_sorted_dh1 = df.sort_values(by='DH1')

# 2. Sắp xếp DH2 tăng dần theo nhóm giới tính
df_sorted_dh2_gt = df.sort_values(by=['GT', 'DH2'])

# 3, 4, 5. Pivot Tables cho DH1
print("Thống kê DH1 theo KT:")
display(df.groupby('KT')['DH1'].agg(stats_list))

print("\nThống kê DH1 theo KT và KV:")
display(df.groupby(['KT', 'KV'])['DH1'].agg(stats_list))
```

### Phần 2 & 3: Trình bày và Trực quan hóa dữ liệu

```python
# --- PHẦN 2 & 3: TRÌNH BÀY DỮ LIỆU ---

# 1. Biểu đồ GT
gt_counts = df['GT'].value_counts()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
gt_counts.plot(kind='bar', ax=ax1, color=['skyblue', 'pink'], title='Tần số Giới tính')
gt_counts.plot(kind='pie', ax=ax2, autopct='%1.1f%%', title='Tần suất Giới tính')
plt.show()

# 4. KV cho Nam, Kinh, đạt điểm sàn
# Lưu ý: 'M' là Nam, DT 1.0 là Kinh dựa trên dữ liệu hiện tại
cond = (df['GT'] == 'M') & (df['DT'] == 1.0) & (df['DH1'] >= 5) & (df['DH2'] >= 4) & (df['DH3'] >= 4)
print("Thống kê KV cho Nam, Kinh, đạt điểm sàn:")
display(df[cond]['KV'].value_counts())
```

### Phần 4 & 5: Phân lớp và Khảo sát phân phối

```python
# --- PHẦN 4 & 5: NÂNG CAO & PHÂN PHỐI ---

# 4.2 Phân lớp T1
df['phanlopt1'] = pd.cut(df['T1'], bins=[0, 5, 7, 8, 10], labels=['k', 'tb', 'kh', 'g'], right=False)

# 5.1 Khảo sát T1
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['T1'], kde=True, ax=ax1)
stats.probplot(df['T1'], dist='norm', plot=ax2)
plt.show()

# 5.5 Ma trận tương quan DH1, DH2, DH3
sns.heatmap(df[['DH1', 'DH2', 'DH3']].corr(), annot=True, cmap='coolwarm')
plt.title('Ma trận tương quan DH1, DH2, DH3')
plt.show()
```
